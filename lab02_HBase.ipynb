###ЗАДАНИЕ

Была создана функция put_many.rb, и добавлен следующий код:

```python
def put_many(table_name, row, column_values)
  column_values.each do |col, val|
    put table_name, row, col, val
  end
end
#Функция принимает список колонок и значений, и добавляет их все в таблицу.
```
